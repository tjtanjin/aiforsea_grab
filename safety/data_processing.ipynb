# Data Processing

### Import libraries

```python
import pandas as pd
import numpy as np
```

Read in filtered data.

```python
combined_df = pd.read_csv("./data/final_filtered.csv")
```

Reduce dimension by calculating the magnitude of acceleration and gyro before dropping the x y z columns.

```python
combined_df['acceleration'] = ((combined_df['acceleration_x'])**2 + (combined_df['acceleration_y'])**2 + (combined_df['acceleration_z'])**2)**(1/2)
```

```python
combined_df['gyro'] = ((combined_df['gyro_x'])**2 + (combined_df['gyro_y'])**2 + (combined_df['gyro_z'])**2)**(1/2)
```

```python
combined_df = combined_df.drop(['acceleration_x', 'acceleration_y', 'acceleration_z', 'gyro_x', 'gyro_y', 'gyro_z', 'Bearing'], axis=1)
```

```python
combined_df.head()
```

Use pivot to obtain the mean/max of the columns for each bookingID and save to csv.

```python
pivot = pd.pivot_table(combined_df, 
                           values = ['Accuracy','second','Speed','acceleration','gyro'],
                           index = ['bookingID'],
                           aggfunc = {'Accuracy': np.mean,
                                      'second': max,
                                      'Speed': max,
                                      'acceleration': max,
                                      'gyro': max
                                     }
                          )
```

```python
pivot.reset_index(inplace=True)
pivot.to_csv("./data/final_processed.csv", index=False)
pivot
```
